# Functions for calculating performance / backtesting

```python
def annual_returns(returns):
    num_years = len(returns) / 252
    cum_ret_final = np.prod(returns + 1)
    return (cum_ret_final ** (1 / num_years) - 1) * 100

def calmar_ratio(daily_pnl):
    # Calculate annual returns
    annual_return_value = annual_returns(daily_pnl)
    
    # Calculate maximum drawdown
    max_drawdown_value = max_drawdown(daily_pnl) / 100  # Convert to percentage
    
    # Calculate Calmar ratio
    if max_drawdown_value != 0:
        calmar = annual_return_value / max_drawdown_value
    else:
        calmar = np.nan  # Avoid division by zero
    
    return calmar

def sharpe(returns, risk_free=0):
    adj_returns = returns - risk_free
    return (np.nanmean(adj_returns) * np.sqrt(252)) / np.nanstd(adj_returns, ddof=1)

def downside_risk(returns, risk_free=0):
    adj_returns = returns - risk_free
    sqr_downside = np.square(np.clip(adj_returns, None, 0))
    return np.sqrt(np.nanmean(sqr_downside) * 252)

def sortino(returns, risk_free=0):
    adj_returns = returns - risk_free
    drisk = downside_risk(returns, risk_free)
    if drisk == 0:
        return np.nan
    return (np.nanmean(adj_returns) * np.sqrt(252)) / drisk

def calculate_winning_percentage(daily_pnl):
    """
    Calculate the winning percentage based on the daily PnL.

    Args:
    - daily_pnl (pd.Series): Series containing the daily PnL.

    Returns:
    - int: winning days
    - int: total days
    - float: The winning percentage.
    """
    # Calculate the number of winning days
    winning_days = int((daily_pnl > 0).sum())
    
    # Total number of days
    total_days = len(daily_pnl)
    
    # Calculate winning percentage
    winning_percentage = (winning_days / total_days) * 100
    
    return winning_days, total_days, winning_percentage

def volatility(returns, annualize=True, trading_days=252):
    """
    Calculate the volatility of a time series of returns.
    
    :param returns: A pandas Series or numpy array of returns.
    :param annualize: Boolean flag to indicate whether to annualize the volatility.
    :param trading_days: The number of trading days in a year (default is 252).
    :return: The calculated volatility.
    """
    # Calculate the standard deviation of returns
    vol = np.std(returns)
    
    # Annualize the volatility if required
    if annualize:
        vol *= np.sqrt(trading_days)
    
    return vol * 100

def max_drawdown(returns):
    wealth_index = 1000 * (1 + returns).cumprod()
    previous_peaks = wealth_index.cummax()
    drawdown = (wealth_index - previous_peaks) / previous_peaks
    return drawdown.min() * 100

def average_drawdown(returns):
    wealth_index = 1000 * (1 + returns).cumprod()
    previous_peaks = wealth_index.cummax()
    drawdown = (wealth_index - previous_peaks) / previous_peaks
    return drawdown[drawdown < 0].mean() * 100 

def calculate_total_trades(spread_position):
    """
    Calculate the total number of trades based on the spread position.

    Parameters:
    - spread_position (array): Array containing the spread positions (-1, 0, 1).

    Returns:
    - total_trades (int): Total number of trades.
    """
    total_trades = 0
    prev_position = None

    for position in spread_position:
        if prev_position is not None and position != prev_position:
            total_trades += 1
        prev_position = position

    return total_trades
```

# OLS Strategy part

```python
def OLS_regression(stock1, stock2):
    # Define the spread and positions for training data
    series1_train = training_data[stock1]
    series2_train = training_data[stock2]
    
    # Perform OLS regression to find hedge ratio
    result = sm.OLS(series1_train, sm.add_constant(series2_train)).fit()
    hedge_ratio = result.params.iloc[1]  # coefficient of stock2 in regression eqn
    intercept = result.params.iloc[0] 
    
    # Calculate the spread
    spread_train = ((series1_train 
                    - hedge_ratio 
                    * series2_train)
                    - intercept)
    
    # Generate trading signals
    spread_mean = spread_train.mean()
    spread_vol = spread_train.std()
    
    # Define the spread and positions for testing data
    series1_test = testing_data[stock1]
    series2_test = testing_data[stock2]
    
    # Calculate the spread for testing data
    spread_test = ((series1_test 
                    - hedge_ratio 
                    * series2_test)
                    - intercept)
    
    # Initialize spread position
    spread_position_test = np.zeros(len(spread_test), dtype=int)
    
    # Define thresholds for spread positions
    threshold1 = 0
    threshold2 = 2
    
    # Round the result of the multiplication to a reasonable precision
    threshold1_condition = np.round(threshold1 * spread_vol, decimals=5)
    threshold2_condition = np.round(threshold2 * spread_vol, decimals=5)
    
    # Assign positions based on spread conditions
    spread_position_test[spread_test > threshold1_condition] = int(-1)  # Short position
    spread_position_test[spread_test < -threshold1_condition] = int(1)  # Long position
    spread_position_test[np.abs(spread_test) > threshold2_condition] = int(0)  # Close out the position
    
    # Shift position by 1 to avoid lookahead bias
    spread_position_test = shift(spread_position_test, 1, cval=0)
    
    # Calculate positions for each stock
    stock1_position_test = spread_position_test
    stock2_position_test = (spread_position_test * hedge_ratio * -1)
    
    # Calculate daily returns
    return_data_test = testing_data.pct_change().shift(-1)  # Daily returns
    
    # Define trading costs
    trading_cost_per_trade = 0.0005  # Example trading cost per trade
    
    # Convert spread_position to a pandas Series
    spread_position_series_test = pd.Series(spread_position_test, index=spread_test.index)
    
    # Calculate trading costs for each trade
    trading_costs_test = spread_position_series_test.diff().abs() * trading_cost_per_trade
    trading_costs_test.iloc[0] = 0  # No trading cost for the first position
    
    # Cumulative trading costs
    cumulative_trading_costs_test = trading_costs_test.cumsum()
    
    # Calculate daily P&L without trading costs
    daily_pnl_without_costs_test = stock1_position_test * return_data_test[stock1] + stock2_position_test * return_data_test[stock2]
    
    # Calculate daily P&L with trading costs
    daily_pnl_test = daily_pnl_without_costs_test - trading_costs_test
    
    # Calculate cumulative P&L with trading costs
    cumulative_pnl_test = daily_pnl_test.cumsum()
    
    # Create a DataFrame to hold the results for the testing period
    result_test_df = pd.DataFrame({
        'spread': spread_test,
        'spread_vol': spread_vol,
        'spread_position': spread_position_test,
        'hedge_ratio': hedge_ratio,
        'stock1_position': stock1_position_test,
        'stock2_position': stock2_position_test,
        'daily_pnl_without_costs': daily_pnl_without_costs_test,
        'trading_costs': trading_costs_test,
        'cumulative_trading_costs': cumulative_trading_costs_test,
        'daily_pnl': daily_pnl_test,
        'cumulative_pnl': cumulative_pnl_test
    })
    
    # Save the result DataFrame for the testing period to a CSV file
    # result_test_df.to_csv('selected_pair_backtest.csv')
    
    spread_beyond_threshold2 = (np.abs(spread_test) > threshold2 * spread_vol)
    # Print the indices where the spread breaks beyond the thresholds
    # print("\nSpread breaks beyond the +/- 2.5 SD threshold:")
    # print(np.where(spread_beyond_threshold2)[0])
    
    dates_beyond_threshold2 = testing_data.index[spread_beyond_threshold2]
    # print("\nDates when the spread breaks beyond the +/- 2.5 SD threshold:")
    # print(dates_beyond_threshold2)
    
    print("\n***** FINANCIAL PERFORMANCE *****\n")

    # Calculate winning percentage
    winning_days, total_days, winning_percentage = calculate_winning_percentage(result_test_df['daily_pnl'])
    print("\nNumber of winning days:", winning_days)
    print("Total number of days:", total_days)
    print(f"Winning percentage: {winning_percentage:.2f}%")
    
    total_trades = calculate_total_trades(result_test_df['spread_position'])
    print("Total number of trades:", total_trades)
    
    ann_returns = annual_returns(result_test_df['daily_pnl'])

    print(f"\nCummulative return: {result_test_df['cumulative_pnl'].iloc[-2]*100:.5f}%")
    print(f"Annual return: {ann_returns:.5f}%")

    print(f"\nSharpe ratio: {sharpe(result_test_df['daily_pnl']):.5f}")
    print(f"Sortino ratio: {sortino(result_test_df['daily_pnl']):.5f}")
    print(f"Max drawdown: {max_drawdown(result_test_df['daily_pnl']):.5f}%")
    # Calculate and print the Calmar ratio
    calmar = calmar_ratio(result_test_df['daily_pnl'])
    print(f"Calmar ratio: {calmar:.5f}")
    
    # Plot cumulative PnL with proper date format on x-axis
    plt.figure(figsize=(10, 6))
    plt.plot(result_test_df.index, result_test_df['cumulative_pnl'].values, label=f"{stock1}, {stock2} Pairs Trading")
    plt.title(f"{stock1}, {stock2} Pairs Trading")
    plt.xlabel("Date")
    plt.ylabel("PnL")
    plt.legend()
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator())
    plt.gcf().autofmt_xdate()  # Rotate date labels
    plt.show()
    
    # Plot spread with shaded areas and proper date format on x-axis
    plt.figure(figsize=(10, 6))
    plt.plot(testing_data.index, spread_test, color='blue', label='Spread')
    plt.axhline(y=0, color='red', linestyle='--', label='Zero')
    plt.axhline(y=-threshold2 * spread_vol, color='green', linestyle='--', label='-2.5 SD Threshold')
    plt.axhline(y=threshold2 * spread_vol, color='green', linestyle='--', label='+2.5 SD Threshold')
    # Shade the regions where the spread exceeds the +/- 2.5 SD threshold
    for date in dates_beyond_threshold2:
        plt.axvline(x=date, color='orange', linestyle='-', alpha=0.5)
    # Shade areas below and above zero line
    plt.fill_between(testing_data.index, spread_test, 0, where=(spread_test <= 0), color='green', alpha=0.5)
    plt.fill_between(testing_data.index, spread_test, 0, where=(spread_test > 0), color='red', alpha=0.5)
    plt.title('Spread Plot with Thresholds and Shaded Exceeding Areas')
    plt.xlabel('Date')
    plt.ylabel('Spread')
    plt.legend()
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator())
    plt.gcf().autofmt_xdate()  # Rotate date labels
    plt.show()
    result_test_df
    return result_test_df
```
